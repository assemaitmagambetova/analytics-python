```python
import pandas as pd

# Загружаем данные
df_audience = pd.read_csv("Данные для тестового задания (1).csv")
df_audience["date"] = pd.to_datetime(df_audience["date"])

# Идентифицируем пользователей, пришедших 1 ноября (День 0)
cohort_users = set(df_audience[df_audience["date"] == "2023-11-01"]["user_id"])
cohort_size = len(cohort_users)

# Находим пользователей, которые были активны 2 ноября (День 1)
day1_users = set(df_audience[df_audience["date"] == "2023-11-02"]["user_id"])

# Считаем пересечение: сколько пользователей из когорты 1 ноября вернулись 2 ноября
returned_day1 = cohort_users.intersection(day1_users)
cnt_returned_day1 = len(returned_day1)

# Вычисляем Retention 1-го дня
retention_day1 = (cnt_returned_day1 / cohort_size) * 100

print(f"Размер когорты от 1 ноября: {cohort_size} пользователей.")
print(f"Вернулись на следующий день (2 ноября): {cnt_returned_day1} пользователей.")
print(f"Retention 1-го дня: {retention_day1:.2f}%")
```

```python
import pandas as pd

# Загружаем данные
df_audience = pd.read_csv("Данные для тестового задания (1).csv")

# 1. Считаем общее число уникальных пользователей за ноябрь
total_users = df_audience["user_id"].nunique()

# 2. Считаем уникальных пользователей, у которых количество просмотров объявлений больше 0
users_with_views = df_audience[df_audience["view_adverts"] > 0][
    "user_id"
].nunique()

# 3. Рассчитываем конверсию в процентах
conversion = (users_with_views / total_users) * 100

print(f"Общее количество уникальных пользователей: {total_users}")
print(
    f"Количество пользователей, посмотревших объявления: {users_with_views}"
)
print(f"Пользовательская конверсия в просмотр: {conversion:.2f}%")
```

```python
import pandas as pd

# Загружаем данные
df_audience = pd.read_csv("Данные для тестового задания (1).csv")

# 1. Считаем суммарное количество всех просмотров за месяц
total_views = df_audience["view_adverts"].sum()

# 2. Считаем количество уникальных пользователей (MAU)
total_users = df_audience["user_id"].nunique()

# 3. Находим среднее количество просмотров на пользователя
average_views_per_user = total_views / total_users

print(f"Общее количество просмотров: {total_views}")
print(f"Всего уникальных пользователей: {total_users}")
print(f"Среднее количество объявлений на пользователя: {average_views_per_user:.2f}")
```

```python
import pandas as pd

# 1. Загружаем ваш файл
df = pd.read_csv("Данные для тестового задания (1).csv")

# 2. Считаем MAU (Уникальные пользователи за месяц)
mau = df["user_id"].nunique()

# 3. Считаем средний DAU (Среднее количество уникальных пользователей в день)
df["date"] = pd.to_datetime(df["date"])
dau_by_day = df.groupby("date")["user_id"].nunique()
average_dau = dau_by_day.mean()

# 4. Считаем пользовательскую конверсию в просмотры
total_users = df["user_id"].nunique()
users_with_views = df[df["view_adverts"] > 0]["user_id"].nunique()
conversion = (users_with_views / total_users) * 100

# 5. Считаем среднее количество просмотров на одного пользователя
total_views = df["view_adverts"].sum()
average_views = total_views / total_users

# Выводим все результаты на экран
print(f"1. MAU продукта: {mau}")
print(f"2. Средний DAU: {average_dau:.2f}")
print(f"3. Конверсия в просмотр объявления: {conversion:.2f}%")
print(f"4. Среднее число просмотров на пользователя: {average_views:.2f}")
```
